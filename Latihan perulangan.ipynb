```python
# input ukuran belah ketupat
n = int(input("Masukkan ukuran belah ketupat: "))

# loop untuk mencetak bagian atas belah ketupat
i = 0
while i < n:
    j = 0
    while j < n-i-1:
        print(" ", end="")
        j += 1
    j = 0
    while j < 2*i+1:
        print("*", end="")
        j += 1
    print()
    i += 1

# loop untuk mencetak bagian bawah belah ketupat
i = n-2
while i >= 0:
    j = 0
    while j < n-i-1:
        print(" ", end="")
        j += 1
    j = 0
    while j < 2*i+1:
        print("*", end="")
        j += 1
    print()
    i -= 1
```
